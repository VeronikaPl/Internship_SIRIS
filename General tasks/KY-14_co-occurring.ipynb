```python
key_words_full_list = list(df["author_key_words"])
```

```python
len(key_words_full_list)
```

```python
dic = {} # the dictionary for the final result
for item in key_words_full_list:
    for keyword in item:
        keys = dic.keys()
        if keyword in keys:
            current_list = dic[keyword]
            for element in item:
                if element != keyword:
                    current_list.append(element)
            dic[keyword] = current_list            
        else:
            current_list = []
            for element in item:
                if element != keyword:
                    current_list.append(element)
            dic[keyword] = current_list                    
```
